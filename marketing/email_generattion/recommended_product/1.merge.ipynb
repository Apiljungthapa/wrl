```python
def fnal(data):
    # Initialize an empty list to store the unique items
    unique_items = []

    for sublist in data:
        # Initialize an empty list for the unique sublist
        unique_sublist = []
        for item in sublist:
            # Only add item if it is not already in the unique_sublist
            if item not in unique_sublist:
                unique_sublist.append(item)
        # Append the unique sublist to the unique_items list
        unique_items.append(unique_sublist)

    return unique_items
```
