```python
from langchain.llms import Ollama
import time
import pandas as pd
```

```python
df=pd.read_excel("raw_products_for_homestead_keywest.xlsx")
```

```python
ollama = Ollama(base_url="http://localhost:11434", model="mymdl")
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
content=df['content'].to_list()
websites=df['website'].to_list()
```

```python
trend_products=[ollama.invoke(raw_text) for raw_text in content]
```

```python
for i in trend_products:
    print(i)
    print("\n")
```

```python
dict={"website":websites,
      "Trending_Products": trend_products}
```

```python
df1=pd.DataFrame(dict)
```

```python
df1
```

```python
for i in zip(df1['website'],df1['Trending_Products']):
    print(i)
```

```python
df1.to_excel("homestead_keywest_trending_products.xlsx", index=False)
```
