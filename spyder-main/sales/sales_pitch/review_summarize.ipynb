## Importing ollama model with langchain

```python
from langchain.llms import Ollama
```

```python


ollama = Ollama(base_url="http://localhost:11434", model="mysum")
```

# Generating the review summary

```python
sum_rev=[ollama.invoke(raw_text) for raw_text in df['Review_Comment']]
```

```python
df=['Summary_Review']=sum_rev
```

```python
df.to_excel("Version_3.xlsx",index=False)
```
