```python
import os
import re
import shutil

# List of names to filter
names_to_keep = ['Puro Cigars Lounge', 'Cigar Bundles of Miami and Lounge', 'Saga Bay Cigars', 'Miami Humidor Cigar Shop US1',
                 'Don Benigno Cigars (Online Store)', 'El Arte del Habano Cigars', 'Cigar Cellar Of Miami',
                 'Miami Humidor Cigar Shop Miller', 'Holy Smokes Kendall', 'Little Havana Cigar Factory', 'Cuban Crafters',
                 'Top Cigars', 'Miami Tobacco Traders - Cigars, Wine, Beer & Art', 'Havana Classic Cigar',
                 'The Lord of the Cigars, Corp.', "Xavier's Cigar Lounge", 'RaceTrac', 'Mega Wine & Spirits',
                 'West Kendall Liquors & Wines', '444 smoke and vape', 'SOFLO SMOKE SHOP', 'Next Level Smoke shop',
                 'Pink Hippo Smoke And Vape', 'Vapes Xpress', 'Mystic Joint Kava Bar & Vape Shop',
                 'The Original Key West Cigar Factory', 'Island Mystique LLC', 'Complete Body Defense Key West', 'Chevron',
                 'Circle K', 'Shell', 'Old Town Wine & Spirits', 'Key West Fine Wines', 'Habibi Corporation']

# Directory containing the review files
directory = 'reviews'

# Directory to store the filtered review files
final_directory = 'final_review'
os.makedirs(final_directory, exist_ok=True)

# Convert names to a regular expression pattern for matching
pattern = re.compile('|'.join([re.escape(name) for name in names_to_keep]), re.IGNORECASE)

# Loop through files in the directory
for filename in os.listdir(directory):
    if pattern.search(filename):
        # If the filename matches any of the names, move the file to the final directory
        source_path = os.path.join(directory, filename)
        destination_path = os.path.join(final_directory, filename)
        shutil.move(source_path, destination_path)
        print(f"Moved: {source_path} to {destination_path}")
    else:
        print(f"Skipped: {filename}")
```
