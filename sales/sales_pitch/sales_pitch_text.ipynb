# making propmt for llm

```python

template = """
{shop} is a {type} shop located at {address} in {city}. 
Their major products featured on their website include {trending_products}.
Here is a summary of their reviews: {summary_review}
"""
```

# combining all the columns of dataframe

```python
formatted_values = []

for _, row in df.iterrows():
    formatted_values.append(template.format(
        shop=row['shop'],
        type=row['type'],
        address=row['address'],
        city=row['City'],
        trending_products=row['Trending_Products'],
        summary_review=row['Summary_Review']
    ))

for formatted in formatted_values:
    print(formatted)
    print("*"*120)
```

```python
df['overall_summary'] = formatted_values
```

```python
df.to_excel("Version_4.xlsx",index=False)
```
